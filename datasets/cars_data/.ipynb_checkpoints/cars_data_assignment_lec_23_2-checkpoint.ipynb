### Building Models

```python
# === display evaluation scores dataframe ===
def get_eval_df(y_val, y_pred, col):
    data = []
    n = len(y_pred)
    k = len(cols)+1
    data.append(metrics.mean_absolute_error(y_val, y_pred))
    data.append(metrics.mean_squared_error(y_val, y_pred))
    data.append(np.sqrt(metrics.mean_squared_error(y_val, y_pred)))
    data.append(metrics.r2_score(y_val, y_pred))
    data.append(1 - (1-metrics.r2_score(y_val, y_pred)) * (n-1)/(n-k-1))
    eval_df = pd.DataFrame(data, ['MAE', 'MSE', 'RMSE', 'R^2', 'R^2_adj'], columns=[col])
    return eval_df

# === Ploting regression line between test and prediction ===
def plt_reg(y_val, y_pred):
    plt.figure()
    sns.regplot(x=y_pred, y=y_val).set(title='Regression plot of Testing and Prediction')
    
# === fitting, predicting ===
def eval_model(X_train, y_train, X_val, y_val, model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    print('Training R^2 score:  ', model.score(X_train, y_train))
    print('Testing R^2 score:   ', model.score(X_val, y_val))
    return y_val, y_pred, model
```
