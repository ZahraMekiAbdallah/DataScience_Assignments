Get the USD To EUR Exchange rate from this web service
https://api.frankfurter.app/latest?amount=1&from=USD&to=EUR
and then use it to make a software that takes amount of USD Dollars from the user and calculate how much will it cost in EUR.
hint: take a look on the url of the service can you guess how I can modify it to calculate the 10$ for EUR for example ?!

```python
def change_amount_currancy(amount=1, from_curr='USD', to_curr='EUR'):
    try:
        response=request('GET','https://api.frankfurter.app/latest?amount='+str(amount)+'&from='+str(from_curr)+'&to='+str(to_curr))
        return ('Amount '+str(amount)+' '+str(from_curr)+' is equal to '+str(response.json()['rates'][to_curr])+' '+str(to_curr))
    except Exception as e:
        return(e)
    
change_amount_currancy()
```

Task 2
---------
Use this web service 
https://www.metaweather.com/api/location/1521894/
one time for 3 different days in this week 
and make a program that get the weather temperature, humidity, visibility, air pressure, wind speed & datetime and save each day data in a CSV & Excel file.

```python
# Return weather of 3 different random days from "consolidated_weather" key
def get_weather_data():
    try:
        response=request('GET','https://www.metaweather.com/api/location/1521894/')
        header = ['Weather temperature', 'Min Temp', 'Max Temp','Humidity', 'Visibility', 'Air Pressure',
                  'Wind Speed', 'Weather Date', 'Created On']
        days_data = random.sample(response.json()['consolidated_weather'], 3)
        data_lst = [[rec['the_temp'], rec['min_temp'], rec['max_temp'], rec['humidity'],
                         rec['visibility'], rec['air_pressure'], rec['wind_speed'],
                         rec['applicable_date'], rec['created']] for rec in days_data]
        df = pd.DataFrame(data_lst, columns=header) 
        df.to_csv('weather_data.csv', encoding='utf-8', index=False)
        df.to_excel('weather_data.xlsx', encoding='utf-8', index=False)
        return df.head()
    except Exception as e:
        return(e)
        
get_weather_data()
```

```python
# Append every time all data from response
def get_weather_data():
    try:
        response=request('GET','https://www.metaweather.com/api/location/1521894/')
        header = ['Weather temperature', 'Min Temp', 'Max Temp','Humidity', 'Visibility', 'Air Pressure',
                  'Wind Speed', 'Weather Date', 'Created On']
        data_lst = [[rec['the_temp'], rec['min_temp'], rec['max_temp'], rec['humidity'],
                         rec['visibility'], rec['air_pressure'], rec['wind_speed'],
                         rec['applicable_date'], rec['created']] 
                    for rec in response.json()['consolidated_weather']]
        
        is_exist = os.path.exists(str(Path().absolute())+'/weather_data.csv')
        if is_exist:
            weather_df=pd.read_csv('weather_data.csv')
            new_df=pd.DataFrame(data_lst, columns=header)
            updated_df = weather_df.append(new_df, ignore_index=True)
            updated_df.to_csv('weather_data.csv', encoding='utf-8', index=False)
            updated_df.to_excel('weather_data.xlsx', encoding='utf-8', index=False)
            return updated_df.head(10)
        else:
            df = pd.DataFrame(data_lst, columns=header) 
            df.to_csv('weather_data.csv', encoding='utf-8', index=False)
            df.to_excel('weather_data.xlsx', encoding='utf-8', index=False)
            return df.head(10)
    except Exception as e:
        return(e)
        
get_weather_data()
```

Task 3
---------
Use this web service 
https://jsonplaceholder.typicode.com/users
and make a program that get all users data including name, username, email, street, suite, city, zip code, geo_lat & geo_long and then save them in a CSV & Excel file.

```python
def get_user_data():
    try:
        response=request('GET','https://jsonplaceholder.typicode.com/users')
        header = ['Name', 'Username', 'Email', 'Street', 'Suite', 'City', 'Zip Code', 'Geo Lat', 'Geo Long']
        data_lst = []
        for rec in response.json():
            data_lst.append([rec['name'],rec['username'],rec['email'],rec['address']['street'],rec['address']['suite'],
                             rec['address']['city'], rec['address']['zipcode'], rec['address']['geo']['lat'], 
                             rec['address']['geo']['lng']])
        df = pd.DataFrame(data_lst, columns=header) 
        df.to_csv('user_data.csv', encoding='utf-8', index=False)
        df.to_excel('user_data.xlsx', encoding='utf-8', index=False)
        return df.head()
    except Exception as e:
        return(e)
        
get_user_data()
```
