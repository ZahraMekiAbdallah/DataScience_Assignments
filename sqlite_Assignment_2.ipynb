#### 7) Write a sql query to add a new category in categories table and add 5 books to this category in books table with title, rate and price.

```python
#=== Add category ===
cursor_obj.execute('insert into category (name) values ("Machine_Learning")')
categ_id = cursor_obj.lastrowid
connection_obj.commit()

#=== Add books ====
books_data=[
    ['Introduction to Machine Learning with Python',800,5,categ_id],
    ['Learning from Data: A Short Course',600,4,categ_id],
    ['Machine Learning for Dummies',700,3,categ_id],
    ['Machine Learning in Action',500,2,categ_id],
    ['Understanding Machine Learning',400,3,categ_id]
]
for book_d in books_data:
    cursor_obj.execute('insert into book (name,price,rate,categ_id) values (?,?,?,?)',book_d)
    connection_obj.commit()

#=== check values ===    
rows=connection_obj.execute('select * from book where categ_id = ?',[categ_id]).fetchall()
rows
```

#### 14) Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.

```python
rows=connection_obj.execute('select bo.name, cat.name, bo.rate, bo.price from book as bo join category as cat where bo.categ_id=cat.id').fetchall()
rows
```

#### 15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.

```python
rows=connection_obj.execute('select rate, count(*) from book where price between 20 and 30 group by rate order by rate').fetchall()
rows
```

#### 16) Write a sql query to calculate how many books each category has having count > 10.

```python
rows=connection_obj.execute('select cat.name, count(*) from book as bo join category as cat where cat.id=bo.categ_id group by categ_id having count(*)>10 order by count(*)').fetchall()
rows
```

#### 17) Write a sql query to get all books with category_name='Music' using subquery.

```python
rows=connection_obj.execute('select * from book where categ_id in (select id from category where name="Music")').fetchall()
rows
```
